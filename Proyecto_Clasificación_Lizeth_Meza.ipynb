# Proyecto de Clasificación
## Data set por usar: "Mushrooms"

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("mushrooms.csv")
df.head()
# Se observan que las columnas son datos de clasificación así que es un buen data set para analizar
```

```python
# Revisamos la estadística descriptiva de los datos e identificamos nuestra variable objetivo, en este data set es "class"
df.describe()
```

```python
df.isna().sum(axis=0)
# Se observa que no se encuentran datos faltantes en el data set y se puede empezar a trabajar con ella
```

## Creamos las matrices 'X' y 'y'

```python
# Definimos "X" y "y"
y=df['class']
df=df.drop(columns=['class'])
X=df
```

```python
# Normalizamos los datos en "y"
le=LabelEncoder()
y=le.fit_transform(y)
y
```

```python
# Hacemos lo mismo con "X" ayudandonos con un ciclo for
for i in X:
    X[i] =le.fit_transform(X[i])
```

```python
# empezamos a separar nuestras database en entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=7487)
```

```python
#Seleccionamos el modelo para clasificación
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.get_params()
```

## Hacemos la búsqueda de los mejores hiperparámetros para ambos modelos.

```python
from sklearn.model_selection import GridSearchCV
```

```python
parametros1={"criterion": ("gini", "entropy"),
              "n_estimators" : (10, 20, 30), 
              "max_samples" : (1/3, 2/3)}
mejor_modelo1=GridSearchCV(model,parametros1,scoring="accuracy")
```

```python
# Entrenamos en busca de los mejores parámetros
mejor_modelo1.fit(X_train,y_train)
print(mejor_modelo1.cv_results_["params"])
```

```python
#Mejores parámetros (seleccionados aleatoriamente)
print(mejor_modelo1.best_score_)
print(mejor_modelo1.best_params_)
```

```python
# Hacemos la predicción con el mejor modelo
y_pred=mejor_modelo1.predict(X_test)
```

```python
# Mandamos a llamar al segundo modelo
from sklearn.naive_bayes import GaussianNB
model2=GaussianNB()
model2.get_params()
```

```python
# Entrenamos el modelo
model2.fit(X_train,y_train)
```

```python
y_pred2=model2.predict(X_test)
```

## Reporte de métricas de ambos modelos

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
# Accuracy del primer modelo
accuracy_score(y_test,y_pred)
```

```python
# Accuracy del segundo modelo
accuracy_score(y_test,y_pred2)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred2)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred2))
```
